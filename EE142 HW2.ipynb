```
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.mlab as mlab
```

Question 1 


```
q = np.single(1e-12)
eps_0 = np.single(8.854e-12)
a = 1e-2

x = np.linspace(-0.2e-2,0.2e-2,2001)
y = x
X, Y = np.meshgrid(x, y)

k = q/(4*np.pi*eps_0)

def V_point(x, y, k, x0, y0):
  V = k/(np.sqrt((x-x0)**2 + (y-y0)**2))
  return V

V_1 = V_point(X, Y, -k, -a, a)
V_2 = V_point(X, Y, k, a, a)
V_3 = V_point(X, Y, k, -a, -a)
V_4 = V_point(X, Y, -k, a, -a)

V = V_1 + V_2 + V_3 + V_4

plt.imshow(V.transpose(), origin='lower', extent = (x[0], x[-1], y[0], y[-1]))
plt.title('Potential')
plt.show()
```

```
E_x, E_y = np.gradient(V)

plt.imshow(E_x.transpose(), origin='lower', extent = (x[0], x[-1], y[0], y[-1]))
plt.title('|E_x| (inside the square)')
plt.show()

plt.imshow(E_y.transpose(), origin='lower', extent = (x[0], x[-1], y[0], y[-1]))
plt.title('|E_y| (inside the square)')
plt.show()
```

```
x = np.linspace(-0.2e-2,.2e-2,21)
y = x
X, Y = np.meshgrid(x, y)

k = q/(4*np.pi*eps_0)

def V_point(x, y, k, x0, y0):
  V = k/(np.sqrt((x-x0)**2 + (y-y0)**2))
  return V

V_1 = V_point(X, Y, -k, -a, a)
V_2 = V_point(X, Y, k, a, a)
V_3 = V_point(X, Y, k, -a, -a)
V_4 = V_point(X, Y, -k, a, -a)

V = V_1 + V_2 + V_3 + V_4

E_x, E_y = np.gradient(V)


plt.quiver(X, Y, E_x, E_y, scale = 0.1)
plt.title('Electric Field Vectors (inside the square)')
plt.show()
```
